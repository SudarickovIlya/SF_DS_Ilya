```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://ru.wikipedia.org/wiki/Города-миллионеры_России'
```

# Способ 1

```python
# Читаем ссылку
response = requests.get(url=url)
# Создаем список всех таблиц на странице
list_of_tables = pd.read_html(response.text)
# Сохраняем нужную
millionaires = list_of_tables[1]
display(millionaires)
```

# Способ 2

```python
# Читаем ссылку
response = requests.get(url=url)
# Считываем страницу в BeautifulSoup
page = BeautifulSoup(response.text, 'html.parser')
# Находим нужную таблицу
table = page.find('table', class_='standard')
# Переводим в формат Pandas.DataFrame
combined_list = []

for row in table('tr'):
    new_list = []
    for cell in row('td'):
        new_list.append(cell.text)
    combined_list.append(new_list)

final_table = pd.DataFrame(combined_list)
```
