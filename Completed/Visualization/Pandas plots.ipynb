```python
import pandas as pd

covid_df = pd.read_csv('data/covid_df.csv', sep=',', decimal='.')
```

```python
grouped_cases = covid_df.groupby('date')['daily_confirmed'].sum()
print(grouped_cases.plot(
    kind='line',
    figsize=(12, 4),
    grid=True,
    title='   ',
    legend=True,
    lw=2
))
```

```python
# -10    

grouped_country = covid_df.groupby('country')['confirmed'].last()
print(grouped_country.head(7))
grouped_country = grouped_country.nlargest(10)
grouped_country.plot(
    kind='bar',
    figsize=(12, 4),
    grid=True,
    legend=True,
    colormap='plasma', 
    title='-10    '
)
```

```python
grouped_country = covid_df.groupby('country')[['confirmed', 'deaths']].last()
grouped_country = grouped_country.nlargest(10, columns=['confirmed'])
grouped_country.plot(
    kind='bar',
    figsize=(12, 4),
    legend=True,
    grid=True,
)
```

```python
covid_df.groupby('country')['total_vaccinations'].last().nsmallest(5).plot(kind='bar', figsize=(12, 3), grid=True, legend=True)
```
