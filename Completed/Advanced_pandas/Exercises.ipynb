```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
movies = pd.read_csv('data/movies.csv', sep = ',', decimal = '.')
ratings1 = pd.read_csv('data/ratings1.csv', sep = ',', decimal = '.')
ratings2 = pd.read_csv('data/ratings2.csv', sep = ',', decimal = '.')
dates = pd.read_csv('data/dates.csv', sep = ',', decimal = '.')
```

```python
ratings = pd.concat([ratings1, ratings2], ignore_index=True)
ratings = ratings.drop_duplicates(ignore_index=True)
ratings
```

```python
rating_dates = pd.concat([ratings, dates], axis=1)
rating_dates
```

```python
# 8.1

merged = rating_dates.merge(movies, on='movieId', how='left')
merged['year_release'] = merged['title'].apply(get_year_release)
merged.info()
```

```python
# 8.2

merged[merged['year_release'] == 1999].groupby(by='title').agg('max').sort_values(by='rating')
```

```python
# 8.3

merged[merged['year_release'] == 2010].groupby(by='genres').mean().sort_values(by='rating')
```

```python
# 8.4

merged.groupby(by='userId')['genres'].nunique().sort_values()
```

```python
# 8.5

merged.groupby('userId', as_index=False)['rating'].agg(['count', 'mean']).sort_values(by=['count', 'mean'], ascending=[True, False])
```

```python
merged[merged['year_release'] == 2018].groupby(by='genres')['rating'].agg(['count', 'mean']).sort_values(by='mean', ascending=False).head(10)
```

```python
# 8.7

merged['year_rating'] = pd.to_datetime(merged['date'], dayfirst=True)
merged['year_rating'] = merged['year_rating'].dt.year
pivot_rating = merged.pivot_table(values='rating', index='genres', columns='year_rating', aggfunc='mean', fill_value=0)
pivot_rating
```
